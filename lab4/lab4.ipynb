## 1. Series
  * Pandas에는 기본 자료구조로 Series와 dataFrame이 있습니다.
  * Series는 1차원 배열과 유사하며 index가 존재합니다.
  * 일반적인 리스트나 배열로부터 만들 수 있습니다.
  * slicing 같은 배열 연산도 가능하며 다양한 메서드(method)가 지원됩니다.
  

```python
import pandas as pd
data = pd.Series([0.25, 0.5, 0.75, 1.0])
data
```

```python
print(data.values)
print(data.index)
print(data[2])
print(data[0:2])
```

## 2. DataFrame
  * DataFrame은 2차원 배열과 유사합니다.
  * dictionary와 밀접한 관련이 있습니다.
  * 생성은 numpy array로부터 생성하는 방법, dictionary로부터 생성하는 방법 등이 있습니다.
  * 그 외에도 다양한 메서드가 지원됩니다.

```python
area = pd.Series({
    "정보기술대": 7,
    "도서관": 6,
    "자연과학대": 5,
    "공과대": 8
})

depart = pd.Series({
    "정보기술대": "컴퓨터공학부",
    "도서관": "열람실",
    "자연과학대": "생명과학부",
    "공과대": "기계공학과"
})

inu = pd.DataFrame({"area": area, "depart": depart})
inu
```

## 3. Operating on Data
  * Series 데이터 사이의 연산은, 기본적으로 index를 기준으로 실행합니다
  * 두 개의 Series 사이의

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
irisdf = pd.DataFrame(iris.data, columns=iris.feature_names)
irisdf['target'] = iris.target
irisdf['target'] = irisdf['target'].map({ 0: "setosa", 1: "versicolor", 2: "virginica"})
irisdf
```

```python
import matplotlib.pyplot as plt
SL = irisdf.iloc[:,0]
SW = irisdf.iloc[:,1]
PL = irisdf.iloc[:,2]
PW = irisdf.iloc[:,3]
name = irisdf.iloc[:,4]
plt.scatter(SL, SW, c=iris.target)
```

```python
plt.scatter(PL, PW, c=iris.target)
```
