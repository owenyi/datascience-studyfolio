### 넘파이 배열 접근  
numpy array 접근 방법은 매우 중요합니다.  
효율적이고 빠르게 사용할 수 있는 데이터를 만드는 것도 중요하지만, 올바르게 사용할 수 있도록 접근하는 일은 더욱 중요합니다.  
  
numpy array의 중요한 속성으로 차원 dimension, 형태 shape, 크기 size가 있습니다.  
이 중 가장 중요한 속성은 차원 dimension 입니다.  
배열 array의 확장된 개념인 텐서 tensor를 이야기 할 때는 랭크 rank 라고도 표현합니다.  
numpy array를 다룰 때는 항상 머리속에 그 array의 차원이 어떤지 그려 본 다음 코딩을 진행해야 합니다.

```python
# numpy array 속성 지정

np.random.seed(0) # seed for reproducibility(재현 불가능), 난수 값 고정

x1 = np.random.randint(10, size=6) # 1차원 array
x2 = np.random.randint(10, size=(3, 4)) # 2
x3 = np.random.randint(10, size=(3, 4, 5)) # 3

print("x3 ndim :", x3.ndim)
print("x3 shape :", x3.shape)
print("x3 size :", x3.size)

print("dtype :", x3.dtype)

print("itemsize :", x3.itemsize, "bytes")
print("nbytes :", x3.nbytes, "bytes")
```

```python
# array 인덱싱

print("x1 :")
print(x1)

print("x1[0] :", x1[0])

print("x1[-2] :", x1[-2])

x1[0] = 3.14159 # 소수점 이하 잘림
print("x1[0] :", x1)

print()

print("x2 :")
print(x2)

print("x2[0, 0] :", x2[0, 0])

print("x2[2, -1] :", x2[2, -1])

x2[0, 0] = 12
print("x2[0, 0] :", x2)
```

```python
# array 슬라이싱 : subarray에 접근
# x[start=0 : stop=<size of dimension> : step=1]

x = np.arange(10)
print(x)

print(x[:5])

print(x[4:7])

print(x[::2])

print(x[::-1])

print(x[5::-2])
```

```python
# 다차원 subarray

print(x2)

print(x2[:2, :3]) # 2행까지, 3열까지

print(x2[:3, ::2]) # 3행까지(모두), 하나 건너 열

print(x2[:, 0]) # 모든 행, 첫 열(indexing) == 첫 열

print(x2[0, :]) # 첫 행
print(x2[0])

x2_sub = x2[:2, :2]
print(x2_sub)

x2_sub[0, 0] = 99
print(x2_sub)

print(x2)
```

### 배열 재구조화  
reshape() 메소드로 array를 재구조화 할 수 있습니다.  
다만 잘 작동하려면 크기를 잘 맞추어야 합니다.  
  
concatenate() 메소드와 split() 메소드를 사용해 array 접합과 분할도 가능합니다.

```python
# array 재구조화

grid = np.arange(1, 10).reshape((3, 3))
print(grid)

x = np.array([1, 2, 3])

print(x.reshape((1, 3))) # reshape을 이용한 행 벡터

r_x = x[np.newaxis, :] # newaxis를 이용한 행 벡터
print(r_x)
print(r_x.shape)

c_x = x[:, np.newaxis]
print(c_x)
print(c_x.shape)
```

```python
# array 접합(concatenation)

# 일차원

x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
print(np.concatenate([x, y]))

z = [99, 99, 99]
print(np.concatenate([x, y, z]))

# 이차원

grid = np.array([[1, 2, 3],
                [4, 5, 6]])
print(np.concatenate([grid, grid])) # np.concatenate([grid, grid], axis=0)
print(np.concatenate([grid, grid], axis=1)) # 두 번째 축을 따라 연결

# 서로 다른 차원

grid = np.array([[9, 8, 7],
               [6, 5, 4]])
x = np.array([1, 2, 3])
y = np.array([[99],
            [99]])

print(np.vstack([x, grid])) # 수직으로
print(np.hstack([grid, y])) # 수평으로
```

```python
# array 분할(splitting)

x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3 = np.split(x, [3, 5])
print(x1, x2, x3)

grid = np.arange(16).reshape((4, 4))
print(grid)

upper, lower = np.vsplit(grid, [2])
print(upper)
print(lower)

left, right = np.hsplit(grid, [2])
print(left)
print(right)
```

### 넘파이 함수
numpy array 계산은 파이썬 array 계산에 비해 매우 빠르게 처리될 수 있습니다.  
속도를 높일 수 있는 핵심은 일반적으로 numpy의 범용 함수(universal functions)를 통해 구현되는 벡터화된 연산을 사용하는 점입니다.  
numpy의 함수는 기본적으로 벡터화된 연산을 사용하므로 반복 계산을 훨씬 더 효율적으로 만드는 데 이용할 수 있습니다.

```python
x = np.arange(4)
print("x =", x)
print("x + 5 =", x + 5)
print("x ** 2 =", x ** 2)
print("x // 2 =", x // 2)
print("x % 2 =", x % 2)
print("-x = ", -x)
```

```python
x = np.array([-2, -1, 0, 1, 2])

print(abs(x))
print(np.absolute(x))
print(np.abs(x))

x = np.array([3 - 4j, 4 -3j, 2 + 0j, 0 + 1j])
print(np.abs(x))
```

```python
theta = np.linspace(0, np.pi, 3)

print("theta      = ", theta)
print("sin(theta) = ", np.sin(theta))
print("cos(theta) = ", np.cos(theta))
print("tan(theta) = ", np.tan(theta))

x = [-1, 0, 1]
print("x         = ", x)
print("arcsin(x) = ", np.arcsin(x))
print("arccos(x) = ", np.arccos(x))
print("arctan(x) = ", np.arctan(x))

x = [1, 2, 3]
print("x     =", x)
print("e^x   =", np.exp(x))
print("2^x   =", np.exp2(x))
print("3^x   =", np.power(3, x))

x = [1, 2, 4, 10]
print("x        =", x)
print("ln(x)    =", np.log(x))
print("log2(x)  =", np.log2(x))
print("log10(x) =", np.log10(x))

x = [0, 0.001, 0.01, 0.1]
print("exp(x) - 1 =", np.expm1(x))
print("log(1 + x) =", np.log1p(x))
```

### (Lab2) 미분 정의를 이용하는 간단한 미분계산  
  
  * f (x)는 함수이고, a는 기울기를 구할 점이며, f’(a)는 점에서의 기울기입니다.
  * 다음 방정식에 대해 Python으로 미분 계산을 코딩합니다.(h 기준 값 0.0000000001 추천)
    * f1(x) = x\*\*2 # for instance f1(2) = 4
    * f2(x) = x\*\*3 + x\*\*2 + x # for f2(2) = ?
    * f3(x) = x\*sin(x\*x\*x+3) # for f3(2) = ?
    * f4(x) = cos(sqrt(e\*\*x + 1) / 2)	# for f4(2) = ?
  1. 코딩의 결과와 이 사이트의 결과를 비교해보세요. https://www.derivative-calculator.net/
  2. f1~f4 함수와 그 도함수의 그래프를 그려보세요.

```python
import numpy as np

h = 0.0000000001

# f1(x)
def f1(x):
    return x**2
fa1 = np.array([2, f1(2)])
print(fa1)

def derf1(x):
    return (f1(x + h) - f1(x)) / h
print(derf1(2))


# f2(x)
def f2(x):
    return x**3 + x**2 + x
fa2 = np.array([2, f2(2)])
print(fa2)

def derf2(x):
    return (f2(x + h) - f2(x)) / h
print(derf2(2))

# f3(x)
def f3(x):
    return x * np.sin(x**3 + 3)
fa3 = np.array([2, f3(2)])
print(fa3)

def derf3(x):
    return (f3(x + h) - f3(x)) / h
print(derf3(2))

# f4(x)
def f4(x):
    return np.cos(np.sqrt((np.exp(x) + 1)) / 2)
fa4 = np.array([2, f4(2)])
print(fa4)

def derf4(x):
    return (f4(x + h) - f4(x)) / h
print(derf4(2))
```
